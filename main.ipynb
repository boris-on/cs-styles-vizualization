```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
```

```python
data
```

```python
data.columns
```

```python
plt.figure(figsize=(8,5))
sns.pairplot(
    data,
    x_vars=['Opening kill ratio', 'Opening kill rating',
       'Team win percent after first kill'],
    y_vars=['Opening kill ratio', 'Opening kill rating',
       'Team win percent after first kill'],
    height = 3,
    aspect = 1.5
)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x="awp", y="Total kills")
plt.text(x = data["awp"][data.nick == 's1mple'],
         y = data["Total kills"][data.nick == 's1mple'],
         s = "s1mple",
         fontdict = dict(color='g', size=8))
sns.scatterplot(data["awp"][data.nick == 's1mple'],
                data["Total kills"][data.nick == 's1mple'], color='g') 
```
